### **add the new columns to the dataframe**

```python
df['genre(s)']=pd.NA
df['explicit']=pd.NA
df['duration-ms']= pd.NA
df['tempo']=pd.NA
df['loudness']=pd.NA
df['energy']=pd.NA
df['danceability']=pd.NA
```

### **get each track's title from base dataset to collect data for the additional columns added**

```python
x=0
#iterating through each song in the dataset
for idx, row in df.iterrows():
    title=row['title']
    performer=row['performer']
    x=x+1
    
    if pd.isnull(row.explicit)==False:
        continue
    
    
    #check cache first 
    if f"{title} {performer}" in songCache:
        df.loc[idx, "explicit"] =songCache[title][performer]['explicit']
        df.loc[idx, "duration-ms"] =songCache[title][performer]['duration_ms']
        df.loc[idx, "tempo"] =songCache[title][performer]['tempo']
        df.loc[idx, "loudness"] =songCache[title][performer]['loudness']
        df.loc[idx, "energy"] =songCache[title][performer]['energy']
        df.loc[idx, "danceability"] =songCache[title][performer]['danceability']
        df.loc[idx, "genre(s)"] =songCache[title][performer]['genres']
        df.to_csv("billboard 2020-2025.csv",index=None)
        continue
    

    #general search
    resp = requests.get(
        f"{URL_BASE}search",
        params={"q": f'{title} {performer}', "type": "track", "limit": 1},
        headers=headers,   
    )

    genData = resp.json()

    ##assigning data from what is collected
    df.loc[idx, "explicit"]=genData['tracks']['items'][0]['explicit']
    df.loc[idx, "duration-ms"]=genData['tracks']['items'][0]['duration_ms']

    #get artist and track ID for further processing
    artistID=genData['tracks']['items'][0]['artists'][0]['id']
    trackID=genData['tracks']['items'][0]['id']


    #collecting track data
    response = requests.get(
        f"https://api.spotify.com/v1/audio-features/{trackID}", headers=headers
    )

    trackData = response.json()

    #assigning track information
    df.loc[idx, "tempo"]=trackData['tempo']
    df.loc[idx, "loudness"]=trackData['loudness']
    df.loc[idx, "energy"]=trackData['energy']
    df.loc[idx, "danceability"]=trackData['danceability']

    if f'{performer}'in artistCache:
        df.loc[idx, "genre(s)"]=songCache[title][performer]['genres']
        df.to_csv("billboard 2020-2025.csv",index=None)
        continue

    #get info from artist
    resp = requests.get(
        f"{URL_BASE}artists/{artistID}",
        headers={"Authorization": "Bearer " + data["access_token"]},   
    )
    
    artistData = resp.json()

    #assigning artist information
    df.loc[idx, "genre(s)"]=','.join(artistData['genres'])
    
    df.to_csv("billboard 2020-2025.csv",index=None)

    if x%100==0:
       print(x)
        
```
